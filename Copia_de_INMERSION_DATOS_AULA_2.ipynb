```python
tipos_seleccionados = ['Oficina/Consultorio', 'Bodega', 'Local']
inmuebles_seleccionados = inmuebles[inmuebles['Tipo'].isin(tipos_seleccionados)]
inmuebles_seleccionados['Tipo'].nunique()
```

```python
inmuebles_seleccionados.sample(10)
```

```python
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_seleccionados, element='bars', x='Precio_Millon', multiple='stack', kde=True, hue='Tipo', palette='bright')
grafica.set_title('Distribucion de inmuebles seleccionados en Bogotá', fontsize=14, fontweight='bold')
plt.xlim(50,2000)
plt.xticks(fontsize=11)
plt.xlabel('Precio/Millon', fontsize=12, fontweight='bold', fontstyle='italic')
plt.yticks(fontsize=11)
plt.ylabel('Cantidad', fontsize=12, fontweight='bold', fontstyle='italic')
nota_explicativa = "Aqui es posible ver los tres tipos seleccionados y la diferencia de cantidades de cada uno segun el precio que esta costando"
plt.annotate(nota_explicativa, xy=(0.5, -0.15), xycoords='axes fraction', fontsize=11, color='gray', va='center', ha='center')
plt.savefig('/content/drive/MyDrive/Ciencia de Datos - AluraLatam/MaterialClase/C2/valor_inmuebles_seleccionados.png',format='png')
plt.show()
```

```python
inmuebles_bodega = inmuebles[inmuebles['Tipo'] == 'Bodega']
inmuebles_casa = inmuebles[inmuebles['Tipo'] == 'Casa']
inmuebles_local = inmuebles[inmuebles['Tipo'] == 'Local']
```

```python
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_bodega, x='Precio_Millon', kde=True, hue='Tipo', palette=['green'])
grafica.set_title('Distribucion de inmuebles de tipo Bodega en Bogotá', fontsize=14, fontweight='bold')
plt.xticks(fontsize=11)
plt.xlabel('Precio/Millon', fontsize=12, fontweight='bold', fontstyle='italic')
plt.yticks(fontsize=11)
plt.ylabel('Cantidad', fontsize=12, fontweight='bold', fontstyle='italic')
plt.savefig('/content/drive/MyDrive/Ciencia de Datos - AluraLatam/MaterialClase/C2/valor_inmuebles_bodega.png',format='png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_casa, x='Precio_Millon', kde=True, hue='Tipo', palette=['red'])
grafica.set_title('Distribucion de inmuebles de tipo Casa en Bogotá', fontsize=14, fontweight='bold')
plt.xticks(fontsize=11)
plt.xlabel('Precio/Millon', fontsize=12, fontweight='bold', fontstyle='italic')
plt.yticks(fontsize=11)
plt.ylabel('Cantidad', fontsize=12, fontweight='bold', fontstyle='italic')
plt.xlim((50,2000))
plt.savefig('/content/drive/MyDrive/Ciencia de Datos - AluraLatam/MaterialClase/C2/valor_inmuebles_casa.png',format='png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_local, x='Precio_Millon', kde=True, hue='Tipo', palette='viridis')
grafica.set_title('Distribucion de inmuebles de tipo Local en Bogotá', fontsize=14, fontweight='bold')
plt.xticks(fontsize=11)
plt.xlabel('Precio/Millon', fontsize=12, fontweight='bold', fontstyle='italic')
plt.yticks(fontsize=11)
plt.ylabel('Cantidad', fontsize=12, fontweight='bold', fontstyle='italic')
plt.savefig('/content/drive/MyDrive/Ciencia de Datos - AluraLatam/MaterialClase/C2/valor_inmuebles_local.png',format='png')
plt.show()
```
